```
import matplotlib.pyplot as plt 
import numpy as np 

G1=[12,7]
G2=[3,5]
G3=[2,7]
u1=[12/16,7/16]
u2=[3/17,5/17]
u3=[2/15,7/15]
x=[]
total_u=[]
mainZ=[]
z=[]
```

```

for i in range(2,100+1):
  Z1=(13*(G1[i-1])+G1[i-2]+3) % 16
  G1.append(Z1)
  u1.append(Z1/16)

print(" ")
print("G1:",G1)  
print("u1:",u1)  

for i in range(2,100+1):
  Z2=(12*(G2[i-1]**2)+13*(G2[i-2])) % 17
  G2.append(Z2)
  u2.append(Z2/17)

print(" ")
print("G2:", G2)  
print("u2:",u2)

for i in range(2,100+1):
  Z3=((G3[i-1]**3)+(G3[i-2])) % 15
  G3.append(Z3)
  u3.append(Z3/15)

print(" ")
print("G3:", G3)
print("u3:",u3)
print(" ")

for i in range(0,100):
  new_u=u1[i]+u2[i]+u3[i]
  
  if (new_u>1):
    edit_u=new_u-1
    total_u.append(edit_u)
  else:
    total_u.append(new_u)
  

# print(z)
print("Total u:",total_u)
print("")
print("For 100:")
for i in range(100):
  xis=i
  x.append(xis)
plt.bar(x,total_u,align='center') # A bar chart
plt.xlabel('i')
plt.ylabel('random number')
```
